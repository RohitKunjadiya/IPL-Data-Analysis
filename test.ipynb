```python
import pandas as pd
```

```python
df = pd.read_csv('ipl_matches.csv')
```

```python
df[df['winning_team'] == 'NR'].shape
```

```python
ipl = df[~df['team1players'].isna()]
```

```python
import ast 

def parse_players(x):
	try:
		if pd.isna(x):
			return []
		return list(ast.literal_eval(x).values())
	except:
		return []

t1players = ipl['team1players'].apply(parse_players)
t2players = ipl['team2players'].apply(parse_players)

team1_match = t1players.apply(lambda players: 'V Kohli' in players)
team2_match = t2players.apply(lambda players: 'V Kohli' in players)

Matches_played = (team1_match | team2_match).sum()
```

```python
import ast
import pandas as pd

def parse_players(x):
    try:
        if pd.isna(x):
            return []

        parsed = ast.literal_eval(x)

        if isinstance(parsed, dict):
            return list(parsed.values())
        elif isinstance(parsed, list):
            return parsed
        else:
            return []
    except Exception as e:
        print("Error:", x, "->", e)
        return []

t1players = ipl['team1players'].apply(parse_players)
t2players = ipl['team2players'].apply(parse_players)

team1_match = t1players.apply(lambda players: 'Virat Kohli' in players)
team2_match = t2players.apply(lambda players: 'Virat Kohli' in players)

Matches_played = (team1_match | team2_match).sum()

print("Matches played:", Matches_played)
```
